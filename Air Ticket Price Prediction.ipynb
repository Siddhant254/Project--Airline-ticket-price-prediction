```python
# IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',50)
```

```python
# READING DATA
df = pd.read_excel('Data_Train.xlsx')
df.head()
```

```python
# DATASET INFO
df.info()
```

```python
# NUMBER OF UNIQUE VALUES IN EACH COLUMN
df.nunique()
```

```python
# DUPLICATES IN DATASET
df.duplicated().sum()
```

```python
# STATISTICAL INFORMATION
df.describe(include=['O']).T
```

```python
# STATISTICAL INFORMATION - PRICE COLUMN
df.describe().T
```

```python
# CHECKING FOR NULL VALUES
df.isnull().sum().sort_values(ascending=False)
```

```python
# UNIQUE VALUES IN EACH COLUMN
for i in df.columns:
    print(f'The unique values in the feature {i} is :',df[i].unique(),sep='\n')
    print()
```

```python
# VALUE COUNTS FOR EACH COLUMN
for i in df.columns:
    print(f'The value counts for each value in feature {i} is :',df[i].value_counts(),sep='\n')
    print()
```
